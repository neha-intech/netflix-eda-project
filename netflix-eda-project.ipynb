```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('netflix_titles.csv')
df.head()
```

```python
print(df.shape)
print(df.info())
print(df.isnull().sum())
```

```python
df.dropna(subset=['rating', 'country'], inplace=True)
df.fillna('Unknown', inplace=True)
print("Cleaning done! Rows remaining:", len(df))
```

```python
df['type'].value_counts().plot(kind='bar', color=['coral', 'steelblue'])
plt.title('Movies vs TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='bar', color='coral')
plt.title('Top 10 Countries with Most Content')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
df['year_added'] = df['date_added'].str.strip().str[-4:]
df['year_added'].value_counts().sort_index().plot(kind='line', marker='o', color='purple')
plt.title('Content Added on Netflix Per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
df['rating'].value_counts().plot(kind='bar', color='steelblue')
plt.title('Content by Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
